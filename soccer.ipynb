```python
def get_popularity(x):
  if x<220:
    return 'Relatively unkmown'
  elif x<600:
    return 'Kind of popular'
  elif x<2000:
    return 'Popular'
  else:
    return 'Super-popular'
```

```python
players.page_views.apply(get_popularity)
```

```python
players['popularity'] = players.page_views.apply(get_popularity)
```
